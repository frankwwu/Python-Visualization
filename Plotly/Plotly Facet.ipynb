```python
import os
import numpy as np
import pandas as pd
import calendar
import datetime
import plotly
import plotly.express as px
print (plotly.__version__)
```

```python
file = 'https://raw.githubusercontent.com/frankwwu/R-Knots/master/Bike%20Sharing/train.csv'
df = pd.read_csv(file)
```

```python
df["date"] = df.datetime.apply(lambda x : x.split()[0])
df["hour"] = df.datetime.apply(lambda x : x.split()[1].split(":")[0])
df["weekday"] = df.date.apply(lambda dateString : calendar.day_name[datetime.datetime.strptime(dateString,"%Y-%m-%d").weekday()])
df["month"] = df.date.apply(lambda dateString : calendar.month_name[datetime.datetime.strptime(dateString,"%Y-%m-%d").month])
df["season"] = df.season.map({1: "Spring", 2 : "Summer", 3 : "Fall", 4 :"Winter" })
df["weather"] = df.weather.map({1: "Clear or cloudy", 2 : "Mist",  3 : "Light Rain or Snow", 4 :"Heavy Rain or Snow" })
```

```python
fig = px.scatter(df, x='hour', y='count', color='temp', size='windspeed', facet_row='workingday', facet_col='weather')
fig.show()
```
