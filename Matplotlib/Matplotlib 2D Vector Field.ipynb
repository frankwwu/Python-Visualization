# Matplotlib 2D Vector Field

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(8, 6))

# Contour Plot
X, Y = np.mgrid[-2:2:100j, -2:2:100j]
Z = X * np.exp(-(X ** 2 + Y ** 2))
levels = np.mgrid[round(np.min(Z - 0.05), 1):round(np.max(Z + 0.05), 1):11j]
cp = plt.contourf(X, Y, Z, cmap='RdYlGn', extend='both', levels=levels)
cb = plt.colorbar(cp, ticks=levels)

# Vector Field
Y, X = np.mgrid[-2:2:20j, -2:2:20j]
U =(1 - 2 * (X ** 2)) * np.exp(-((X ** 2) + (Y ** 2)))
V = -2 * X * Y * np.exp(-((X ** 2) + (Y ** 2)))
speed = np.sqrt(U ** 2 + V ** 2)
UN = U / speed
VN = V / speed
quiv = plt.quiver(X, Y, UN, VN, color='dodgerblue', headlength=7)

plt.show()
```

```python
# Creating coordinates
x = np.arange(-8, 8, 0.1)
y = np.arange(-6, 6, 0.1)

# Creating a meshgrid
X, Y = np.meshgrid(x, y)

# Assigning vector directions
Ex = (X + 1) / ((X + 1) ** 2 + Y ** 2) - (X - 1) / ((X - 1) ** 2 + Y ** 2)
Ey = Y / ((X + 1) ** 2 + Y ** 2) - Y / ((X - 1) ** 2 + Y ** 2)

# Depict illustration
plt.figure(figsize=(8, 6))
plt.streamplot(X, Y, Ex, Ey, density=1.2, linewidth=None, color='dodgerblue')
plt.plot(-1, 0, '-or', markersize=10)
plt.plot(1, 0, '-og', markersize=10)
plt.title('Electromagnetic Dipole Field')

plt.grid()
plt.show()
```
