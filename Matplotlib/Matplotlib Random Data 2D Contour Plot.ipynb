```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.tri as tri
import numpy as np
```

```python
np.random.seed(20)
npts = 200
x1 = np.random.uniform(-2, 2, npts)
y1 = np.random.uniform(-2, 2, npts)
z1 = x1 * np.exp(-x1 ** 2 - y1 ** 2)
```

```python
# Directly supply the irregularly spaced coordinates to tricontour.

def tricontour_plot(x, y, z):    
    fig, ax = plt.subplots()     
    levels = 10
    ax.tricontour(x, y, z, levels = levels, linewidths = 0.5, colors='k')
    cntr = ax.tricontourf(x, y, z, levels = levels, cmap = plt.cm.rainbow)
    fig.colorbar(cntr, ax = ax)
    ax.plot(x, y, 'ko', ms = 3, alpha = 0.3)
    ax.set(xlim = (-2, 2), ylim = (-2, 2))
    ax.set_title('tricontour (%d points)' % len(x))     
    plt.show()
    
    
tricontour_plot(x1, y1, z1)
```
