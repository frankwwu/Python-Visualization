```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
dataframe = pd.read_csv('https://raw.githubusercontent.com/frankwwu/R-Knots/master/EEG%20Brain%20Wave%20for%20Confusion/EEG-data.csv')
```

```python
dataframe.head()
```

```python
plt.figure(figsize=(18,15))
cor = dataframe.corr()
sns.heatmap(cor, annot=True, cmap='coolwarm', fmt='.1g')
plt.show()
```
