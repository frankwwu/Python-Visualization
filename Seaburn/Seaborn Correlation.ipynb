```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_csv('https://raw.githubusercontent.com/frankwwu/R-Knots/master/EEG%20Brain%20Wave%20for%20Confusion/EEG-data.csv')
```

```python
dataframe.head()
```

```python
plt.figure(figsize=(18,15))
cor = dataframe.corr()
hm = sns.heatmap(cor, annot=True, cmap='coolwarm', fmt='.1g')
hm.invert_xaxis()
plt.show()
```

```python
plt.figure(figsize=(18,15))
cor = dataframe.corr()
hm = sns.heatmap(cor, annot=True, cmap='coolwarm', fmt='.1g', mask = np.tril(cor.corr()))
hm.invert_xaxis()
#hm.invert_yaxis()
hm.xaxis.set_label_position('top') 
plt.show()
```
