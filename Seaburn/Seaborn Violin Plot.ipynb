```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load example data
tips = sns.load_dataset("tips")
tips.head()
```

```python
ax = sns.violinplot(x="day", y="total_bill", data=tips, inner=None, split=True, horizontal=True)
ax = sns.swarmplot(x="day", y="total_bill", data=tips, color="white", edgecolor="gray")
#ax = sns.stripplot(x="day", y="total_bill", data=tips, color="white", edgecolor="gray")

# Set labels and title
ax.set(xlabel='Day of Week', ylabel='Total Bill', title='Total Bill')

plt.yticks(rotation = 90)

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
palette = sns.color_palette("pastel")

ax = sns.violinplot(y="day", x="total_bill", data=tips, split=True, palette=[palette[0], palette[1], palette[2], palette[3]])

# Set labels and title
ax.set(xlabel='Total Bill', ylabel='Day of Week', title='Total Bill')

plt.yticks(rotation = 90)
plt.show(ax)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

sns.set_style('white')
palette = sns.color_palette("Set2")
palette=[palette[0], palette[1], palette[2], palette[3]]
ax = sns.violinplot(y="day", x="total_bill", data=tips, scale="width", inner=None, palette=palette)
xlim = ax.get_xlim()
ylim = ax.get_ylim()
for violin in ax.collections:
    bbox = violin.get_paths()[0].get_extents()
    x0, y0, width, height = bbox.bounds
    violin.set_clip_path(plt.Rectangle((x0, y0), width, height / 2, transform=ax.transData))
    violin.set_edgecolor('lightgrey')
    
sns.boxplot(y="day", x="total_bill", data=tips, saturation=1, showfliers=False,
            width=0.1, boxprops={'zorder': 3, 'facecolor': 'none'}, ax=ax)
old_len_collections = len(ax.collections)
sns.stripplot(y="day", x="total_bill", data=tips, palette=palette, jitter=0.2, orient='h', ax=ax)
for dots in ax.collections[old_len_collections:]:
    dots.set_offsets(dots.get_offsets() + [0, 0.25])
    
# Set labels and title
ax.set(xlabel='Total Bill', ylabel='Day of Week', title='Total Bill')

ax.set_xlim(xlim)
ax.set_ylim(ylim)
plt.show()
```
