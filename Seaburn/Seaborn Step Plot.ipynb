```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
url = 'https://gist.githubusercontent.com/frankwwu/d48c704aa9c712b51d20c50b9cf5ca07/raw/48f1f95764c5fca12faf167ba2f95edb2f88eaed/CanadaIncomeTaxRates2020.csv'
df = pd.read_csv(url)
```

```python
df['Federal tax rates'] = df['Federal tax rates'].str.rstrip('%').astype(float)
df['British Columbia tax rates'] = df['British Columbia tax rates'].str.rstrip('%').astype(float)
df['Alberta tax rates'] = df['Alberta tax rates'].str.rstrip('%').astype(float)
df['Ontario tax rates'] = df['Ontario tax rates'].str.rstrip('%').astype(float)
df['New Brunswick tax rates'] = df['New Brunswick tax rates'].str.rstrip('%').astype(float)
df['Nova Scotia tax rates'] = df['Nova Scotia tax rates'].str.rstrip('%').astype(float)
```

```python
df
```

```python
ax = sns.set(rc={'figure.figsize':(16, 10)})
ax = sns.lineplot(x=df['Federal tax bracket'], y=df['Federal tax rates'], drawstyle='steps-pre', label='Federal')
ax = sns.lineplot(x=df['British Columbia tax bracket'], y=df['British Columbia tax rates'], drawstyle='steps-pre', label='British Columbial')
ax = sns.lineplot(x=df['Alberta tax bracket'], y=df['Alberta tax rates'], drawstyle='steps-pre', label='Alberta')
ax = sns.lineplot(x=df['Ontario tax bracket'], y=df['Ontario tax rates'], drawstyle='steps-pre', label='Ontario')
ax = sns.lineplot(x=df['New Brunswick tax bracket'], y=df['New Brunswick tax rates'], drawstyle='steps-pre', label='New Brunswick')
ax = sns.lineplot(x=df['Nova Scotia tax bracket'], y=df['Nova Scotia tax rates'], drawstyle='steps-pre', label='Nova Scotia')
ax.set(xlabel="Tax ($)", ylabel = "Rate (%)")
ax.xaxis.set_major_formatter(ticker.EngFormatter())
ax.yaxis.set_major_formatter(ticker.PercentFormatter())
```
