```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = sns.load_dataset('iris')
```

```python
data.head()
```

```python
g = sns.PairGrid(data, vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], hue="species", 
                 diag_sharey=True, palette='husl')
g.map_diag(sns.kdeplot)
g.map_lower(sns.kdeplot, fill=True, levels=5, alpha=.5)
g.map_upper(plt.scatter, alpha=.5)
g.add_legend(title="iris", adjust_subtitles=True)
```

```python
g = sns.PairGrid(data, vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], hue="species", 
                 diag_sharey=False, palette='pastel')
g.map_diag(sns.kdeplot)
g.map_lower(sns.kdeplot, fill=True, levels=5, alpha=.5)
g.map_upper(plt.scatter, alpha=.6)
g.add_legend(title="iris", adjust_subtitles=True)
```

```python
g = sns.PairGrid(data, vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], hue="species", 
                 diag_sharey=False, palette='tab10')
g.map_diag(sns.kdeplot)
g.map_lower(sns.kdeplot, fill=True, levels=5, alpha=.5)
g.map_upper(plt.scatter, alpha=.4)
g.add_legend(title="iris", adjust_subtitles=True)
```
